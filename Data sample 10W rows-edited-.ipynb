```python
def w_avg(df, stars_review, user_weight):
    d = df[stars_review]
    w = df[user_weight]
    
    return (d * w).sum() / w.sum()
```

```python
star_after_weight = sample_user_tier_100000[['business_id','review_id', 'stars_review', 'user_weight']].groupby(by = ['business_id'], as_index=False).apply(lambda x: x)
star_after_weight
```

```python
star_after_weight = sample_user_tier_100000[['business_id','review_id', 'stars_review','user_weight']].groupby(by = ['business_id'], as_index=False).apply(w_avg, 'stars_review', 'user_weight')
star_after_weight
```
